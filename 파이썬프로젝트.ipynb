```python
cols = [
    "channelId",
    "videoCategoryId",
    "subscriberCount",
    "channelViewCount",
    "likes/subscriber",
    "views/subscribers",
    "VideoCommentCount"
]

df = pd.read_csv("./ydc.csv", usecols=cols + ["videoPublished"], encoding="cp949")

# datetime 변환 및 날짜만 추출
df["videoPublished"] = pd.to_datetime(df["videoPublished"], errors="coerce")
df["published_date"] = df["videoPublished"].dt.date

# 불필요한 컬럼 제거
df = df.drop(columns=["videoPublished"])

# published_date 맨 앞으로 컬럼 순서 재정렬
new_order = ["published_date"] + cols
df = df[new_order]
df
```

```python
df.sort_values(by='published_date', ascending=False, inplace=True)
df
```

```python
import pandas as pd
import numpy as np

# 좋아요 비율 변환
df["구독자당 좋아요 비율"] = pd.to_numeric(df["구독자당 좋아요 비율"], errors="coerce")
df["구독자당 좋아요 비율"] = df["구독자당 좋아요 비율"].replace(-1, np.nan)

# 조회수 비율 변환
df["구독자당 조회수 비율"] = pd.to_numeric(df["구독자당 조회수 비율"], errors="coerce")
df["구독자당 조회수 비율"] = df["구독자당 조회수 비율"].replace(-1, np.nan)
```

```python
import numpy as np
import pandas as pd

# 'likes/subscriber' 변환
df["likes/subscriber"] = pd.to_numeric(df["likes/subscriber"], errors="coerce")
df["likes/subscriber"] = df["likes/subscriber"].replace(-1, np.nan)

# 'views/subscribers' 변환
df["views/subscribers"] = pd.to_numeric(df["views/subscribers"], errors="coerce")
df["views/subscribers"] = df["views/subscribers"].replace(-1, np.nan)

# 확인
print(df[["likes/subscriber", "views/subscribers"]].dtypes)
print(df[["likes/subscriber", "views/subscribers"]].head())
```

```python
import matplotlib as mpl
from matplotlib import font_manager

# ① 사용할 수 있는 한글 폰트 중 하나 자동 선택
preferred_fonts = [
    "Malgun Gothic",      # Windows
    "NanumGothic",        # Linux (많이 사용)
    "Noto Sans CJK KR",   # Google Noto
    "Noto Sans KR",       # 또 다른 Noto 이름
    "나눔고딕", "맑은 고딕",
]
available = {f.name for f in font_manager.fontManager.ttflist}
chosen = next((f for f in preferred_fonts if f in available), None)

if chosen is None:
    print("⚠️ 한글 폰트를 찾지 못했습니다. OS에 한글 폰트를 설치한 뒤 다시 실행하세요.")
else:
    mpl.rcParams["font.family"] = chosen
    mpl.rcParams["axes.unicode_minus"] = False  # 음수 기호 깨짐 방지
    print(f"✅ 사용 폰트: {chosen}")
```

```python
rename_dict = {
    "published_date": "게시일",
    "channelId": "채널 ID",
    "videoCategoryId": "동영상 카테고리 ID",
    "subscriberCount": "구독자 수",
    "channelViewCount": "채널 조회수",
    "likes/subscriber": "구독자당 좋아요 비율",
    "views/subscribers": "구독자당 조회수 비율",
    "VideoCommentCount": "동영상 댓글 수",
}
df = df.rename(columns=rename_dict)
df
```

```python
print(df.columns.tolist())
```

```python
cols = ["구독자 수", "채널 조회수", "구독자당 좋아요 비율", "구독자당 조회수 비율", "동영상 댓글 수"]

corr = df[cols].corr(method="pearson")

print(corr)

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1,
            cbar_kws={"label": "상관계수"})
plt.title("채널 지표 상관관계 히트맵")
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 라벨 이름 줄바꿈 적용
df = df.rename(columns={
    "구독자당 좋아요 비율": "구독자당\n좋아요 비율",
    "구독자당 조회수 비율": "구독자당\n조회수 비율",
    "동영상 댓글 수": "동영상\n댓글 수"
})

cols = ["구독자 수", "채널 조회수", "구독자당\n좋아요 비율", "구독자당\n조회수 비율", "동영상\n댓글 수"]
corr = df[cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1,
            cbar_kws={"label": "상관계수"})
plt.title("채널 지표 상관관계 히트맵")
plt.xticks(rotation=0)   # 줄바꿈된 글씨는 rotation 안 해도 잘 보임
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
